```python
question = df_evals.iloc[0]['Question']
course = df_evals.iloc[0]['Course']

number_of_course_rating_questions = 0
number_of_workscore_questions = 0

sum_of_course_rating = 0
sum_of_workscore = 0

data_list = []

for row in df_evals.iterrows():
    change_question_flag = False
    if row[1]['Course'] == course:
        if not math.isnan(row[1]['Course rating']):
            sum_of_course_rating += row[1]['Course rating']
            
        if not math.isnan(row[1]['Workscore']):
            sum_of_workscore += row[1]['Workscore']
            
        if row[1]['Question'] != question and not math.isnan(row[1]['Course rating']):
            number_of_course_rating_questions += 1
            change_question_flag = True
            
        if row[1]['Question'] != question and not math.isnan(row[1]['Workscore']):
            number_of_workscore_questions += 1
            change_question_flag = True
            
        if change_question_flag:
            question = row[1]['Question']
    else:
        data_list.append([course, sum_of_course_rating, sum_of_workscore, number_of_course_rating_questions, number_of_workscore_questions])
        course = row[1]['Course']
        
        if math.isnan(row[1]['Course rating']):
            sum_of_course_rating = 0
        else:
            sum_of_course_rating = row[1]['Course rating']
            
        if math.isnan(row[1]['Workscore']):
            sum_of_workscore = 0
        else:
            sum_of_workscore = row[1]['Workscore']
            
        if math.isnan(row[1]['Course rating']):
            number_of_course_rating_questions = 0
        else:
            number_of_course_rating_questions = 1
            
        if math.isnan(row[1]['Workscore']):
            number_of_workscore_questions = 0
        else:
            number_of_workscore_questions = 1
```

```python
df_course_pairs = pd.read_csv(course_info_csv)

for i, row in df_course_pairs.iterrows():
    cell_name = row['Name'].replace("\n", "")
    regex_catch = re.search(r'<a.*>(.*)</a>', cell_name)
    if regex_catch != None:
        df_course_pairs.at[i, 'Name'] = regex_catch.group(1)
        
    cell_info = row['Info'].replace("\n", "")
    cell_info = cell_info.replace("<div>", " ")
    cell_info = cell_info.replace("</div>", " ")
    cell_info = cell_info.replace("<br>", " ")
    cell_info = cell_info.replace("  ", " ")
    cell_info = cell_info.replace("  ", " ")
    cell_info = cell_info.replace("  ", " ")
    cell_info = cell_info.strip()
    df_course_pairs.at[i, 'Info'] = cell_info
```

```python
elective_table = data_df.merge(df_grades, how='left', on='Course')
elective_table = elective_table[['Course', 'Course rating', 'Workscore', 'Registered', 'Grade', 'Passed', 'Percentage']]

elective_table['Grade'] = elective_table['Grade'].fillna(elective_table['Percentage'] * 12)

elective_table['Value'] = ((elective_table['Course rating'] / 5 * 100) + (elective_table['Workscore'] / 5 * 100) + (elective_table['Grade'] / 12 * 100) + (elective_table['Percentage'] * 100)) / 4

elective_table.sort_values(by='Value', ascending=False).head(15)
```

```python
results = elective_table[(elective_table['Registered'] > 6) & (elective_table['Workscore'] > 3)].sort_values(by='Value', ascending=False)

results
```

```python
expanded_results = results.copy(deep=True)

expanded_results['English title'] = ""
expanded_results['ECTS'] = np.nan
expanded_results['Schedule'] = ""
expanded_results['Duration'] = ""
expanded_results['Evaluation'] = ""

for index, row in results.iterrows():
    course_info_df = df_course_pairs[df_course_pairs['Course'] == row['Course']]
    
    if course_info_df.empty:
        continue
    
    language = course_info_df[course_info_df['Name'] == 'Language of instruction']['Info'].values[0]
    if language != "English":
        continue
    
    name = course_info_df[course_info_df['Name'] == "English title"]['Info'].values[0]
    ects = course_info_df[course_info_df['Name'] == "Point( ECTS )"]['Info'].values[0].replace(",", ".")
    schedule = course_info_df[course_info_df['Name'] == "Schedule"]['Info'].values[0]
    duration = course_info_df[course_info_df['Name'] == "Duration of Course"]['Info'].values[0]
    evaluation = course_info_df[course_info_df['Name'] == "Evaluation"]['Info'].values[0]
    
    expanded_results.at[index, 'English title'] = name
    expanded_results.at[index, 'ECTS'] = float(ects)
    expanded_results.at[index, 'Schedule'] = schedule
    expanded_results.at[index, 'Duration'] = duration
    expanded_results.at[index, 'Evaluation'] = evaluation
```

### Searching good courses

```python
expanded_results = expanded_results[~expanded_results['Course'].isnull()]
expanded_results['Course'] = expanded_results['Course'].astype(int)
filtered_expanded_results = expanded_results[expanded_results["English title"] != ""]

filtered_expanded_results.sort_values(by='Value', ascending=False).head(10)
```
